```python
import os
import pandas as pd
import numpy as np
from scipy.io import loadmat
```

```python
# Write path and file

path_header = "../../osf_files/lsl_physio_data/"
file  = "ID_001_lsl_data.mat"

path = os.path.join(path_header, file)

data = loadmat(path, squeeze_me = True, struct_as_record = False)

streams = [c for c in data['lsl_data']]
print(streams)
```

```python
for stream in data['lsl_data']:
    if stream.info.name in ["pupil_capture", "BrainVision RDA"]:
        labels = [c.label for c in stream.info.desc.channels.channel]
        print(stream.info.name, "->", labels)
```
